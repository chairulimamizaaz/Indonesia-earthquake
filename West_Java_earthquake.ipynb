```python
import pandas as pd
data_gempa1 = '/content/drive/MyDrive/dataset/katalog_gempa.csv'
data = pd.read_csv(data_gempa1)
```

```python
print("Data sebelum membersihkan:")
display(data)
```

```python
data.sample(10)
```

```python
data.info()
```

```python
data['tgl'] = pd.to_datetime(data['tgl'])
data['ot'] = pd.to_datetime(data['ot'])
```

```python
data['year'] = data['tgl'].dt.year
data['moon'] = data['tgl'].dt.month_name()
data['day'] = data['tgl'].dt.day_name()
data['time'] = data['ot'].dt.strftime('%H:%M:%S.%f').str[:-3]
```

```python
kolom_tidak_diperlukan = ['strike1', 'dip1', 'rake1', 'strike2', 'dip2', 'rake2']
data_cleaned = data.drop(columns=kolom_tidak_diperlukan)
```

```python
data_cleaned.info()
```

```python
data = data_cleaned[['tgl', 'time', 'day', 'moon', 'year', 'depth', 'mag', 'remark', 'lat', 'lon']]
display(data)
```

```python
print("Nilai Null dalam DataFrame:")
print(data.isnull().sum())
print("\nJumlah Baris Duplikat dalam DataFrame:", data.duplicated().sum())
```

```python
data = data.drop_duplicates()
```

```python
data[data.duplicated()]
```

```python
# data = data_no_dup
data.rename(columns = {'remark': 'Location', 'mag': 'Magnitude', 'depth': 'Depth (km)', 'tgl': 'Date', 'lat': 'Latitude', 'lon': 'Longitude'}, inplace = True)
data
```

```python
data.info()
```

```python
data_jawa = data[data['Location'].str.contains('Java - Indonesia')]
data_jawa
```

```python
def get_level(magnitude):
    if magnitude < 2.5:
        return 1
    elif magnitude < 5.4:
        return 2
    elif magnitude < 6.0:
        return 3
    elif magnitude < 6.9:
        return 4
    elif magnitude < 7.9:
        return 5
    else:
        return 6
    # else:
    #     return None

data['Level'] = data['Magnitude'].apply(get_level)
data
```

```python
correlation = data['Magnitude'].corr(data['Depth (km)'])
correlation
```

Korelasi antara Magnitude dan Depth (km) sebesar 0.2356415062203313 menunjukkan adanya korelasi positif yang lemah antara kedua variabel tersebut. Artinya, ada kecenderungan bahwa semakin besar magnitudo gempa bumi, semakin dalam kedalaman gempa tersebut cenderung terjadi, meskipun korelasi tersebut tidak terlalu kuat.

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
plt.scatter(data['Magnitude'], data['Depth (km)'], alpha=0.5)
plt.title('Korelasi antara Magnitude dan Depth (km)')
plt.xlabel('Magnitude')
plt.ylabel('Depth (km)')
plt.grid(True)
plt.show()
```

```python
import seaborn as sns

plt.figure(figsize=(8, 6))
sns.regplot(x='Magnitude', y='Depth (km)', data=data, scatter_kws={'alpha':0.5})
plt.title('Korelasi antara Magnitude dan Depth (km)')
plt.xlabel('Magnitude')
plt.ylabel('Depth (km)')
plt.grid(True)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

corr_matrix = data[['Magnitude', 'Depth (km)']].corr()

plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Heatmap Korelasi antara Magnitude dan Depth (km)')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

numeric_columns = data.select_dtypes(include=['float64', 'int64'])
corr_matrix = numeric_columns.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Heatmap Korelasi')
plt.show()
```
