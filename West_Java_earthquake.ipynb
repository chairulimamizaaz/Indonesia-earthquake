```python
import pandas as pd
data_gempa1 = '/content/drive/MyDrive/dataset/katalog_gempa.csv'
data = pd.read_csv(data_gempa1)
```

```python
print("Data sebelum membersihkan:")
display(data)
```

```python
data.sample(10)
```

```python
data.info()
```

```python
data['tgl'] = pd.to_datetime(data['tgl'])
data['ot'] = pd.to_datetime(data['ot'])
```

```python
data['tahun'] = data['tgl'].dt.year
data['bulan'] = data['tgl'].dt.month_name()
data['hari'] = data['tgl'].dt.day_name()
data['waktu'] = data['ot'].dt.strftime('%H:%M:%S.%f').str[:-3]
# data = data.drop(columns=['ot'])
```

```python
kolom_tidak_diperlukan = ['strike1', 'dip1', 'rake1', 'strike2', 'dip2', 'rake2']
data_cleaned = data.drop(columns=kolom_tidak_diperlukan)
```

```python
data_cleaned.info()
```

```python
data = data_cleaned[['tgl', 'waktu', 'hari', 'bulan', 'tahun', 'depth', 'mag', 'remark', 'lat', 'lon']]
display(data)
```

```python
print("Nilai Null dalam DataFrame:")
print(data.isnull().sum())
print("\nJumlah Baris Duplikat dalam DataFrame:", data.duplicated().sum())
```
